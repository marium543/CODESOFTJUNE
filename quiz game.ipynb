```python
import random

# Define a pool of quiz questions
all_quiz_questions = [
    {
        "question": "What is the capital of France?",
        "options": ["A. Paris", "B. London", "C. Berlin"],
        "correct_answer": "A"
    },
    {
        "question": "Who is the king of animals?",
        "options": ["A. Bear", "B. Lion", "C. Tiger"],
        "correct_answer": "B"
    },
    {
        "question": "Which planet is known as the Red Planet?",
        "options": ["A. Venus", "B. Mars", "C. Jupiter"],
        "correct_answer": "B"
    },
     {
        "question": "Which city is known as The Eternal City?",
        "options": ["A. Rome", "B. New York", "C. Chicago"],
        "correct_answer": "A"
    },
     {
        "question": "How many colors are there in a rainbow?",
        "options": ["A. 6", "B. 8", "C. 7"],
        "correct_answer": "C"
    },
     {
        "question": "Which planet is known as the biggest planet?",
        "options": ["A. Venus", "B. Mars", "C. Jupiter"],
        "correct_answer": "C"
    },
     {
        "question": "Where is the Eiffel Tower?",
        "options": ["A. Paris", "B. Canada", "C. Australia"],
        "correct_answer": "A"
    },
     {
        "question": "Which country has the largest population?",
        "options": ["A. France", "B. China", "C. Germany"],
        "correct_answer": "B"
    }
]

# Function to randomly select three questions for the user
def select_quiz_questions():
    return random.sample(all_quiz_questions, 3)

# Function to display a question and get user's answer
def ask_question(question_data):
    print(question_data["question"])
    for option in question_data["options"]:
        print(option)
    user_answer = input("Your answer (enter the letter): ").strip().upper()
    return user_answer

# Function to check if the user's answer is correct
def check_answer(user_answer, correct_answer):
    if user_answer == correct_answer:
        print("Correct!")
        return True
    else:
        print(f"Incorrect. The correct answer is {correct_answer}.")
        return False

# Main game loop
def play_quiz():
    global score
    selected_quiz_questions = select_quiz_questions()
    
    print("Welcome to the Quiz Game!")
    print("Rules:")
    print("1. Answer the questions by entering the letter corresponding to your choice.")
    print("2. You will get one point for each correct answer.")
    
    for question_data in selected_quiz_questions:
        user_answer = ask_question(question_data)
        if check_answer(user_answer, question_data["correct_answer"]):
            score += 1
    
    print(f"Your Final Score: {score}/{len(selected_quiz_questions)}")
    
    play_again = input("Do you want to play again? (yes/no): ").strip().lower()
    if play_again == "yes":
        score = 0
        play_quiz()
    else:
        print("Thank you for playing!")

# Initialize score
score = 0

# Start the game
play_quiz()
```
